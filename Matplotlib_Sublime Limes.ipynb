```python
import matplotlib.pyplot as plt
import pandas as pd
```

import Python libraries Matplotlib and Pandas

```python
df=pd.read_csv("../input/sublime/Sublime Limes.csv")
```

This dataset is about a make-up online lime retailer called Sublime Limes. The purpose of this project is to gain insight into the customers and their sales habits, and using Matplotlib, to communicate this information effectively.

```python
ax1=plt.subplot()
x_values= range(len(df.months))
plt.plot(x_values, df.visits_per_month, marker='*')
plt.xlabel('Month')
plt.ylabel('Total Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(df.months)
plt.title('Monthly Visits')
```

Plot total page visits over the past year in a line chart

```python
ax2=plt.subplot()
plt.plot(x_values, df.key_limes_per_month, color='red')
plt.plot(x_values, df.persian_limes_per_month, color='green')
plt.plot(x_values, df.blood_limes_per_month, color='gray')
plt.legend(['key limes', 'persian limes', 'blood limes'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(df.months)
plt.title('number of limes by species')
```

plot numbers of limes sold by species over the past year
