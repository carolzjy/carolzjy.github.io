```python
import pandas as pd
```

```python
visits = pd.read_csv('/kaggle/input/Visits.csv')

cart = pd.read_csv('/kaggle/input/Cart.csv')

checkout = pd.read_csv('/kaggle/input/Checkout.csv')

purchase = pd.read_csv('/kaggle/input/Purchase.csv')

print(visits.head())
print(cart.head())
print(checkout.head())
print(purchase.head())
```

The data files are about a made-up company called Cool T-Shirts. The purpose is to analyze data on visits to the website and understand the funnel better.

In this case, The funnel is going to describe the following process:

1. A user visits CoolTShirts.com
2. A user adds a t-shirt to their cart
3. A user clicks “checkout”
4. A user actually purchases a t-shirt

```python
visits_cart= pd.merge(visits, cart, how='left')

all_count=len(visits_cart)

null= visits_cart[visits_cart.cart_time.isnull()]

null_count=len(null)

no_cart= float(null_count)/all_count*100

print(no_cart)
```

Combined visits and cart using a left merge, and created a new dataframe called visits_cart to further understand the conversion from visiting the site to placing items in cart.The result shows 80.5% of users who visited Cool T-Shirts ended up not placing a t-shirt in their cart. 

```python
cart_checkout= pd.merge(cart,checkout, how='left')

null2= cart_checkout[cart_checkout.checkout_time.isnull()]

all_count2= len(cart_checkout)
null_count2= len(null2)

no_checkout= float(null_count2)/all_count2*100

print(no_checkout)
```

Repeat the same process to merge cart and checkout files. 20.9% of users who placed items in their cart did not proceed to checkout. 

```python
all_data= visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')

checkout_to_purchase= all_data[all_data.checkout_time.notnull()]

total_checkouts= len(checkout_to_purchase)

null3= checkout_to_purchase[checkout_to_purchase.purchase_time.isnull()]

total_non_purchase= len(null3)

no_purchase= float(total_non_purchase)/total_checkouts*100

print(no_purchase)
```

Continue to repeat the process and calculate the conversion from checkout to purchase. Only 14.7% of users who proceeded to checkout did not end up making purchases. Comparing all steps of the funnel,the pagevisit s the weakest with the highest percentage of users dropping out. 

```python
all_data['time_to_purchase'] = int(all_data.purchase_time) - int(all_data.visit_time)

print(all_data.time_to_purchase)

print(all_data.time_to_purchase.mean())
```
